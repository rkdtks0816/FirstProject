## 1. 라이브러리 및 데이터
## Library & Data

#### 외부 데이터
- SMP, supply, 기온 데이터들을 2010년 부터 2020년의 데이터를 수집하여 2020-05-18 까지의 데이터만 사용하였습니다.
- SMP(제주) 일별 데이터 [2010-01-01 ~ 2020-05-18] :  https://www.kpx.or.kr/www/contents.do?key=226
- supply(제주) 일별 데이터 [2010-01-01 ~ 2020-05-18] :   https://www.kpx.or.kr/www/contents.do?key=356
- 기온(제주) 일별 데이터(평균, 최저, 최고) [2010-01-01~2020-05-18] : https://data.kma.go.kr/data/grnd/selectAsosRltmList.do?pgmNo=36
- 외부 데이터를 다운로드 받는 시점에 따라 추가적으로 데이터가 더 생기기 때문에 2020-05-18 까지의 데이터만 슬라이싱 하는 과정이 달라 질 수 있음.


```python
import warnings
warnings.filterwarnings('ignore')


import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np
import lightgbm as lgb
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import seaborn as sns
from tqdm.notebook import tqdm


smp_files = glob.glob('data/smp/*')
weather_files=glob.glob('data/jeju_weather/*')
supply_files = glob.glob('data/supply/*')
```

## 2. 데이터 전처리
## Data Cleansing & Pre-Processing  

#### supply
- supply 외부 데이터에는 일별 설비용량, 공급능력, 전년최대전력, 금년최대전력, 공급예비력, 예비율이 존재한다.
- 중간에 비어있는 값이 한 곳 존재하여 앞의날과 뒤의날의 값의 평균으로 채운다.
- 최종적으로 2010-01-01 ~ 2020-05-18 의 데이터를 생성한다.


```python
supply_dfs=[]
for file in supply_files:
    tmp = pd.read_excel(file,encoding='cp949').iloc[1:,1:]
    tmp.columns = ['일시', '설비용량', '공급능력', '전년최대전력', '금년최대전력','증가율','공급예비력','예비율']
    tmp['설비용량'] = tmp['설비용량'].astype(np.float32)
    tmp['공급능력'] = tmp['공급능력'].astype(np.float32)
    tmp['전년최대전력'] = tmp['전년최대전력'].astype(np.float32)
    tmp['금년최대전력'] = tmp['금년최대전력'].astype(np.float32)
    tmp['공급예비력'] = tmp['공급예비력'].astype(np.float32)
    tmp['예비율'] = tmp['예비율'].astype(np.float32)
    tmp = tmp.iloc[::-1]
    supply_dfs.append(tmp)
    
supply_df = pd.concat(supply_dfs)                                                # 불러온 데이터들을 한 데이터로 연결
supply_df['일시'] = [k[:-5].replace(' ','') for k in supply_df['일시'].values]
supply_df['일시'] = pd.to_datetime(supply_df['일시'].values)                     # 날짜 데이터 타입으로 변경
supply_df.reset_index(inplace=True, drop=True)                                  
start = '2010-01-01'
end = '2020-05-18'
tmp = pd.DataFrame(columns = ['일시'])
tmp['일시'] = pd.date_range(start, end, freq='D')            
supply_df = pd.merge(tmp,supply_df,how='outer')                        


# 비어 있는 데이터를 앞의 날과 뒤의 날의 평균으로 채움
supply_df['설비용량'].fillna(102.93,inplace=True)                                 
supply_df['공급능력'].fillna(99.550003,inplace=True)
supply_df['전년최대전력'].fillna((55.750000+55.360001)/2,inplace=True)
supply_df['금년최대전력'].fillna((58.980000+58.349998)/2,inplace=True)
supply_df['증가율'].fillna((5.79+5.40)/2,inplace=True)
supply_df['공급예비력'].fillna((40.570000+41.200001)/2,inplace=True)
supply_df['예비율'].fillna((68.790001+70.610001)/2,inplace=True)

# 최종적으로 5월 18일까지의 데이터를 사용하기 위해 슬라이싱 다운로드 받은 시점으로 부터 슬라이싱 범위를 조절해줄 필요가 있음
supply_df = supply_df[:-4]
```

#### SMP
- SMP 외부 데이터에는 일별 SMP의 최소, 최대, 평균 데이터가 존재한다.
- 비어있는 값이 없으므로 그대로 불러온후 2020-01-01 ~ 2020-05-18의 데이터를 생성

```python
smp_dfs = []
for i in range(len(smp_files)):
    tmp = pd.read_excel(smp_files[i], skiprows=3).iloc[:][['구분','최소','최대','평균']]
    tmp['구분'] = pd.to_datetime(tmp['구분'].astype(str))
    smp_dfs.append(tmp)
    
smp_df = pd.concat(smp_dfs)

# 최종적으로 5월 18일까지의 데이터를 사용하기 위해 슬라이싱 다운로드 받은 시점으로 부터 슬라이싱 범위를 조절해줄 필요가 있음
smp_df = smp_df.iloc[:-1]
```

#### weather
- 제주도의 기온을 외부데이터로 사용하였다. 일별로 지점별 평균, 최저, 최고 기온 데이터가 존재한다.
- 비어있는 값이 없으므로 그대로 불러온다.
- 제주도의 평균적인 기온을 사용하기 위해 날짜별로 지점별 평균, 최저, 최고 기온을 평균내어 사용한다.
- 제주도 기온 2020-01-01 ~ 2020-05-18의 데이터를 생성

```python
weather_dfs = []
for i in range(len(weather_files)):
    tmp = pd.read_csv(weather_files[i],encoding='cp949')
    tmp.columns = ['지점','지점명','구분','평균기온','최저기온','최고기온']
    tmp.drop(['지점명'],axis=1, inplace=True)
    weather_dfs.append(tmp)
    
weather_df = pd.concat(weather_dfs)
temp_mean = weather_df.groupby(['구분']).mean()['평균기온'].values
temp_max = weather_df.groupby(['구분']).mean()['최고기온'].values
temp_min = weather_df.groupby(['구분']).mean()['최저기온'].values
```

#### 최종으로 사용할 데이터를 모두 합쳐 구축

```python
train=pd.DataFrame()
train['date'] = smp_df['구분']
train['smp_min'] = smp_df['최소']
train['smp_max'] = smp_df['최대']
train['smp_mean'] = smp_df['평균']
train['temp_mean'] = temp_mean
train['temp_max'] = temp_max
train['temp_min'] = temp_min
train['facility_capacity'] = supply_df['설비용량'].values
train['supply_ability'] = supply_df['공급능력'].values
train['pre_supply'] = supply_df['전년최대전력'].values
train['supply'] = supply_df['금년최대전력'].values
train['supply_capacity'] = supply_df['공급예비력'].values
train['cap_ratio'] = supply_df['예비율'].values
train['month'] = train['date'].dt.month
train['day'] = train['date'].dt.day
train['dayofweek'] = train['date'].dt.dayofweek
train.reset_index(drop=True, inplace=True)
```

## 4. 변수 선택 및 모델 구축
## Feature Selection & Initial Modeling  
#### feature 선택
- target value인 supply, smp(mean, max, min)을 예측하기 위해 각 입력 데이터들을 다르게 설정하였다.
- supply는 분기별 특성이 뚜렷히 보이는 것 같아 날짜 관련 데이터를 모두 사용하고 supply 관련 데이터, 기온데이터를 모두 사용하였다.
- smp_mean은 smp_mean 과 temp_mean을 사용
- smp_max는 smp_max, smp_mean, 날짜 관련 데이터, temp_mean을 사용
- smp_min은 smp_min, temp_mean, supply 데이터를 사용
- 4개의 파라미터 옵션들로 target들을 각각 예측한다. 이 때 사용하는 파라미터들은 target에 따라 다르다.

### 기상 속성 선정 과정
- 사용한 지점(4곳) & 기상속성(기온)
    - 제주, 고산, 성산, 서귀포의 평균, 최고, 최저 기온 데이터
- 해당 지점과 기상속성을 선정한 이유
    - 성산(동), 고산(서), 서귀포(남), 제주(북)의 네곳의 지점의 평균을 내면 제주도에서 측정된 기상 데이터의 평균적인 값을 알 수 있을 것이고, 기상 속성 중 기온의 데이터만 사용한 이유는 기온은 supply 데이터 처럼 주기적인 특성이 강하기 때문에 supply 데이터 예측에 도움이 될 것이라 생각하였다. 또 결측값이 거의 없기 때문에 기온 데이터만 선정하였다.

```python
# target별 사용한 컬럼
supply_use_col = ['month', 'dayofweek', 'day',
       'temp_mean', 'temp_max', 'temp_min', 'facility_capacity',
       'supply_ability', 'pre_supply', 'supply', 'supply_capacity',
       'cap_ratio']
smp_mean_use_col = ['smp_mean','temp_mean']                           
smp_max_use_col = ['smp_max','day','dayofweek','month','temp_mean', 'smp_mean']
smp_min_use_col = ['smp_min','temp_mean','supply']
```

```python
# 4개의 파라미터
params1 = {
    'objective': 'mae',
    'feature_fraction':0.1, 
}
params2 = {
    'objective': 'mae',
}
params3 = {
    'objective': 'mse',
    'feature_fraction':0.1, 
}
params4 = {
    'objective': 'mse',
}

def TRAIN_lgbm(params, X_train_, y_train_, X_valid_, y_valid_, verbose=0):
    train_data = lgb.Dataset(X_train_, y_train_)
    valid_data = lgb.Dataset(X_valid_, y_valid_)
    model = lgb.train(params, train_data, 100000, (train_data, valid_data), verbose_eval=verbose, early_stopping_rounds=100)
    return model
```

## 5. 모델 학습 및 검증
## Model Tuning & Evaluation

#### 훈련 데이터, 검증데이터 생성
- 30일 동안의 시계열 데이터를 통해 7일 후 ~ 35일 후 (총 28일)을 예측하도록 데이터를 구성한다.
    - 0~29 --> 36
    - 1~30 --> 37
    - 2~31 --> 38
    - ...
- 구성된 데이터 중 최근 20% 데이터를 검증 데이터로 설정하고 나머지는 훈련데이터로 설정한다.
- 훈련 시 100번의 iteration 동안 개선이 없을 시 훈련을 중단한다.
- 한 파라미터 당 7일 후 ~ 35일 후를 예측하기 때문에 28개의 모델들이 나온다.
- 4개의 파라미터를 모두 쓰는 target들은 28개 * 4 로 112개의 모델이 나온다.


### Training - supply

```python
supply_models={}
interval=30
visualization=True

for param_idx, params in enumerate([params1, params2, params3, params4]):
    supply_models_tmp=[]
    for day in tqdm(range(7, 35)):

        X_train, y_train = [], []
        for k in range(0, len(train)-interval-day):
            X = train[supply_use_col].values[k:k+interval]          
            y = train['supply'].values[k+interval+day-1]
            
            X_train.append(X)
            y_train.append(y)
        
        X_train, y_train = np.array(X_train), np.array(y_train)

        valid_idx = -int(len(train)*0.2)

        X_train, X_valid, y_train, y_valid = X_train[:valid_idx], X_train[valid_idx:], y_train[:valid_idx], y_train[valid_idx:]
        
        X_train = X_train.reshape(X_train.shape[0],-1)
        X_valid = X_valid.reshape(X_valid.shape[0],-1)
        
        supply_lgbm_model = TRAIN_lgbm(params, X_train, y_train, X_valid, y_valid, verbose=5000)
        supply_models_tmp.append(supply_lgbm_model)
        print('{}번 param {}day completed'.format(param_idx+1, day))
        
        
        # visualization 을 원할 시 상단의 visualization 옵션을 True로 하면 됨.
        if visualization==True:
            plt.figure(figsize=(20,5))
            plt.plot(y_valid, label='true')
            plt.plot(supply_lgbm_model.predict(X_valid), label='lgb_pred')
            plt.title('{} days later'.format(day))
            plt.legend()
            plt.show()
        
    supply_models[param_idx]=supply_models_tmp
```

### Training - smp_mean
